# Практическое задание №2 по теме "Создание признакового пространства".

Загрузка данных:

```python
import pickle
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA

with open('/content/drive/MyDrive/data.pickle', 'rb') as f:
    df = pickle.load(f)

df.head()
```

Переводим списки в столбцах tweet_stemmed и tweet_lemmatized в str:

```python
for i in range(len(df['tweet_stemmed'])):
    out_str = " "
    strr = df['tweet_stemmed'][i]
    df['tweet_stemmed'][i] = out_str.join(strr)
    out_str = " "
    strr = df['tweet_lemmatized'][i]
    df['tweet_lemmatized'][i] = out_str.join(strr)
```

Машок слов с помощью CountVectorizer:

```python
count_vectorizer = CountVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')

bag_of_words = count_vectorizer.fit_transform(df['tweet_stemmed'])

feature_names = count_vectorizer.get_feature_names_out()
BoW_Cv_ts = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
BoW_Cv_ts.head()
```

```python
bag_of_words = count_vectorizer.fit_transform(df['tweet_lemmatized'])

feature_names = count_vectorizer.get_feature_names_out()
BoW_Cv_tl = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
BoW_Cv_tl.head()
```

Мешок слов с помощью TfidfVectorizer:

```python
tfidf_vectorizer = TfidfVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')
values = tfidf_vectorizer.fit_transform(df['tweet_stemmed'])

feature_names = tfidf_vectorizer.get_feature_names_out()
BoW_Tf_ts=pd.DataFrame(values.toarray(), columns = feature_names)
BoW_Tf_ts.head()
```

```python
values = tfidf_vectorizer.fit_transform(df['tweet_lemmatized'])

feature_names = tfidf_vectorizer.get_feature_names_out()
BoW_Tf_tl=pd.DataFrame(values.toarray(), columns = feature_names)
BoW_Tf_tl.head()
```

Проверка на корпусе:

```python
# Загружаем данные
data = open('/content/drive/MyDrive/corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

Разделение данных:

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
count_vectorizer = CountVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')
tfidf_vectorizer = TfidfVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')
```

Построение моделей:

```python
count_vectorizer.fit(trainDF['text'])

xtrain_count =  count_vectorizer.transform(train_x)
xvalid_count =  count_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions_CV = classifier.predict(xvalid_count)
```

```python
tfidf_vectorizer.fit(trainDF['text'])

xtrain_count =  tfidf_vectorizer.transform(train_x)
xvalid_count =  tfidf_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions_TF = classifier.predict(xvalid_count)
```

```python
print(f'accuracy_score count_vectorizer №1: {accuracy_score(valid_y, predictions_CV)}')
print(f'accuracy_score tfidf_vectorizer №1: {accuracy_score(valid_y, predictions_TF)}')
```

Уменьшим количество признаков:

```python
count_vectorizer = CountVectorizer(max_df = 0.9, max_features = 500, stop_words='english')
tfidf_vectorizer = TfidfVectorizer(max_df = 0.9, max_features = 500, stop_words='english')

count_vectorizer.fit(trainDF['text'])

xtrain_count =  count_vectorizer.transform(train_x)
xvalid_count =  count_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions_CV_2 = classifier.predict(xvalid_count)

tfidf_vectorizer.fit(trainDF['text'])

xtrain_count =  tfidf_vectorizer.transform(train_x)
xvalid_count =  tfidf_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions_TF_2 = classifier.predict(xvalid_count)

print(f'accuracy_score count_vectorizer №1: {accuracy_score(valid_y, predictions_CV)}')
print(f'accuracy_score tfidf_vectorizer №1: {accuracy_score(valid_y, predictions_TF)}')
print(f'accuracy_score count_vectorizer features = 500: {accuracy_score(valid_y, predictions_CV_2)}')
print(f'accuracy_score tfidf_vectorizer features = 500: {accuracy_score(valid_y, predictions_TF_2)}')
```

Как видно из результатов, уменьшение количества признаков отрицательно повлияло на accuracy.

```python
count_vectorizer = CountVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')
tfidf_vectorizer = TfidfVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')

pca = PCA(n_components=500)

count_vectorizer.fit(trainDF['text'])

xtrain_count =  count_vectorizer.transform(train_x)
xvalid_count =  count_vectorizer.transform(valid_x)
xtrain_PCA =  pca.fit_transform(xtrain_count.toarray())
xvalid_PCA =  pca.fit_transform(xvalid_count.toarray())

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_PCA, train_y)
predictions_CV_3 = classifier.predict(xvalid_PCA)

tfidf_vectorizer.fit(trainDF['text'])

xtrain_count =  tfidf_vectorizer.transform(train_x)
xvalid_count =  tfidf_vectorizer.transform(valid_x)
xtrain_PCA =  pca.fit_transform(xtrain_count.toarray())
xvalid_PCA =  pca.fit_transform(xvalid_count.toarray())

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_PCA, train_y)
predictions_TF_3 = classifier.predict(xvalid_PCA)

print(f'accuracy_score count_vectorizer №1: {accuracy_score(valid_y, predictions_CV)}')
print(f'accuracy_score tfidf_vectorizer №1: {accuracy_score(valid_y, predictions_TF)}')
print(f'accuracy_score count_vectorizer features = 500: {accuracy_score(valid_y, predictions_CV_2)}')
print(f'accuracy_score tfidf_vectorizer features = 500: {accuracy_score(valid_y, predictions_TF_2)}')
print(f'accuracy_score count_vectorizer PCA: {accuracy_score(valid_y, predictions_CV_3)}')
print(f'accuracy_score tfidf_vectorizer PCA: {accuracy_score(valid_y, predictions_TF_3)}')
```

Как видно из результатов, применение PCA для уменишения размерности значительно ухудшило первоначальную точность.
